```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("universal_top_spotify_songs.csv")
```

```python
# Display unique values for each column
for col in df.columns:
    unique_values = df[col].unique()
    print(f"Column: {col}")
    print(f"Unique Values: {unique_values[:10]}")  # Show only the first 10 unique values to avoid long outputs
    print("-" * 50)
```

```python
# Data Cleaning: Drop rows with missing values
df_clean = df.dropna()
```

```python
sorted_df_clean = df_clean.sort_values(by='artists')
#sorted_df_clean.head()
sorted_df_clean[['artists']]
```

```python
# Extract relevant columns
df_clean = df_clean[['name', 'artists', 'country', 'popularity']]
```

```python
# Convert 'popularity' to numeric (if it's not already)
df_clean['popularity'] = pd.to_numeric(df_clean['popularity'], errors='coerce')
```

```python
# Visualization: Top 10 most popular songs globally
top_songs = df_clean.groupby('name').agg({'popularity': 'sum'}).reset_index()
top_songs = top_songs.sort_values('popularity', ascending=False).head(10)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='popularity', y='name', data=top_songs, palette='viridis')
plt.title('Top 10 Most Popular Songs Globally')
plt.xlabel('Popularity')
plt.ylabel('Song')
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Clean and prepare data
df_clean = df.dropna(subset=['popularity', 'country', 'name'])
df_clean['snapshot_date'] = pd.to_datetime(df_clean['snapshot_date'])
```

```python
# Example 1: Choropleth Map for popularity by country
choropleth_map = px.choropleth(df_clean, locations="country", color="popularity", hover_name="country",
                                title="Song Popularity by Country")
choropleth_map.update_layout(autosize=True)
```

```python
# Example 2: Line Graph for trends over time
line_graph = px.line(df_clean, x='snapshot_date', y='popularity', color='name',
                     title='Popularity Over Time')
line_graph.update_layout(autosize=True)

```

```python
# Example 3: Bar Chart for popularity by audio feature (danceability example)
bar_chart = px.bar(df_clean, x='name', y='danceability', color='popularity',
                   title="Popularity by Danceability")
bar_chart.update_layout(autosize=True)

```

```python
# Example 4: Scatter Plot for correlation between two features (e.g., Energy vs Danceability)
scatter_plot = px.scatter(df_clean, x='energy', y='danceability', color='popularity',
                          title="Correlation Between Energy and Danceability")
scatter_plot.update_layout(autosize=True)

# Show all visualizations
choropleth_map.show()
line_graph.show()
bar_chart.show()
scatter_plot.show()
```
