# **Column Descriptions**

Here's a description of some of the key columns in the dataset:

*   **artist:** The name of the artist.
*   **song:** The name of the song.
*   **duration_ms:** The duration of the song in milliseconds
*   **explicit:** Whether the song is explicit (1) or not (0).
*   **year:** The year the song was released.
*   **popularity:** The popularity score of the song (0-100).
*   **danceability:** How suitable a track is for dancing (0.0-1.0).
*   **energy:** A measure of intensity and activity (0.0-1.0).
*   **key:** The key the track is in.
*   **loudness:** The overall loudness of a track in decibels (dB).
*   **mode:** Mode indicates the modality (major or minor) of a track.
*   **speechiness:** Detects the presence of spoken words in a track (0.0-1.0).
*   **acousticness:** A confidence measure of whether the track is acoustic (0.0-1.0).
*   **instrumentalness:** Predicts whether a track contains no vocals (0.0-1.0).
*   **liveness:** Detects the presence of an audience in the recording (0.0-1.0).
*   **valence:** A measure of musical positiveness conveyed by a track (0.0-1.0).
*   **tempo:** The overall estimated tempo of a track in beats per minute (BPM).
*   **genre:**The genre of the song.

# Top Genres by Song Count

**Step 2: Top Genres by Song Count**

We then visualize the most represented genres in the dataset:

```python
df['genre'].value_counts().head(10).plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Number of Songs')
plt.title('Top 10 Genres by Song Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

**Step 3: Cleaning Genre Data**

Upon visualizing the top genres, we observed that some songs have multiple genres. To address this, we clean and process the genre data:

```python
df['genre_list'] = df['genre'].str.split(',')

# Explode the DataFrame to create a new row for each genre in the 'genre_list'
df_exploded = df.explode('genre_list')

# Strip leading/trailing spaces from the genre names
df_exploded['genre_list'] = df_exploded['genre_list'].str.strip()

genre_counts = df_exploded['genre_list'].value_counts().reset_index()
genre_counts.columns = ['genre', 'count']  # Rename columns for clarity

# Create the bar graph
fig = px.bar(genre_counts, x='genre', y='count', title='Genre Distribution')
fig.update_xaxes(tickangle=45, tickfont=dict(size=10))  # Rotate x-axis labels
fig.show()
```

This produces a more cohesive visualization by separating genres properly.

# Relationship between Danceability and Energy

**Step 4: Relationship between Danceability and Energy**

We examine the relationship between danceability and energy in songs:

```python
df.plot(kind='scatter', x='danceability', y='energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.title('Relationship between Danceability and Energy')
plt.show()
```

However, the initial visualization is a bit vague. To improve this, we visualize the relationship with respect to genres:

```python
df_exploded = df.assign(genre_list=df['genre'].str.split(',')).explode('genre_list')
df_exploded['genre_list'] = df_exploded['genre_list'].str.strip()  # Remove leading/trailing spaces

# Create the scatter plot with all genres in one plot
import plotly.express as px

# Assuming df_exploded is already created

fig = px.scatter(df_exploded,
                 x='danceability',
                 y='energy',
                 color='genre_list',
                 title='Relationship between Danceability and Energy by Genre',
                 labels={'danceability': 'Danceability', 'energy': 'Energy'},
                 hover_data=['song', 'artist'])

# Add interactive features:
fig.update_layout(
    updatemenus=[
        dict(
            type="dropdown",
            direction="down",
            buttons=list([
                dict(
                    args=[{"visible": [True] * len(df_exploded)},
                          {"title": "All Genres"}],
                    label="All Genres",
                    method="update"
                ),
                *[dict(
                    args=[{"visible": [(genre == g) for genre in df_exploded['genre_list']]},
                          {"title": f"Genre: {g}"}],
                    label=g,
                    method="update"
                ) for g in df_exploded['genre_list'].unique()]
            ]),
        )
    ],
    autosize=False,  # Disable autosizing for better control
    width=1500,  # Set width of the figure
    height=1000   # Set height of the figure
)

fig.show()
```

This updated visualization adds interactivity, allowing users to filter by genre and better understand the relationship between danceability and energy across genres.

NOTE: While using the plotly library for visualizations, warning suggests that, in future versions of pandas, we will need to adjust how we group data when passing a length-1 list-like object. For now, it does not affect the current functionality of the code.

# Average Popularity by Genre

To visualize the average popularity of songs by genre, we initially created a bar chart based on the dataset:

```python
df.groupby('genre')['popularity'].mean().plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.title('Average Popularity by Genre')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

However, we noticed that this visualization doesn't fully capture the nuances, especially since the genre column contains multiple genres for some songs. To address this, we split the genre column and exploded the DataFrame to handle multiple genres:

```python
# Split the genre column and explode the DataFrame
df_exploded = df.assign(genre_list=df['genre'].str.split(',')).explode('genre_list')
df_exploded['genre_list'] = df_exploded['genre_list'].str.strip()  # Remove leading/trailing spaces

# Group by the exploded genre list and calculate average popularity
genre_popularity = df_exploded.groupby('genre_list')['popularity'].mean().reset_index()

# Sort by average popularity in descending order
genre_popularity = genre_popularity.sort_values(by=['popularity'], ascending=False)

# Create the bar chart using Plotly
fig = px.bar(genre_popularity, x='genre_list', y='popularity',
             title='Average Popularity by Genre',
             labels={'genre_list': 'Genre', 'popularity': 'Average Popularity'})

# Customize layout for better readability
fig.update_layout(
    xaxis_tickangle=-45,  # Rotate x-axis labels
    xaxis_title_font=dict(size=14),
    yaxis_title_font=dict(size=14),
    title_font=dict(size=16)
)

fig.show()
```

This approach produces a much clearer visualization and helps us better understand the average popularity of each genre.

**Visualizing Trends in Audio Features Over Time**

For the next step, we wanted to explore how different audio attributes, such as danceability, energy, and loudness, have changed over time. To do this, we decided to include interactivity so that we could visualize trends per attribute over the years.

We defined the list of attributes we wanted to analyze. Then, we calculated the average value of each attribute per year. And to make the visualization interactive, we used a line plot where users could select an attribute from a dropdown menu to view its trend over time:

```python
attributes = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
              'instrumentalness', 'liveness', 'valence', 'tempo', 'popularity']

# Calculate average values per year for each attribute
avg_values_per_year = df.groupby('year')[attributes].mean().reset_index()

fig = px.line(avg_values_per_year, x='year', y='danceability',
              title="Average Attribute Value per Year",
              labels={'year': 'Year', 'value': 'Average Value'})

fig.update_traces(mode='lines+markers', line=dict(width=2), marker=dict(size=6))

fig.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=[{'y': [avg_values_per_year[attr]], 'label': attr}],
                    label=attr,
                    method="update"
                ) for attr in attributes
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top"
        ),
    ],
    xaxis_title="Year",
    yaxis_title="Average Value",
    font=dict(size=14),
    template="plotly_white"
)

fig.show()
```

This interactive plot allows users to explore the trends of each attribute over the years. By selecting different attributes from the dropdown menu, users can visualize how each feature (e.g., danceability, energy, popularity) has evolved from 2000 to 2019. The interactivity enhances the experience by allowing users to focus on specific attributes of interest, giving them the flexibility to dive deeper into the data.
